# Exploratory Data Analysis






# Gender


```python
fig = sns.countplot(x = 'Gender', data = df)

for bars in fig.containers:
    fig.bar_label(bars)
plt.title("Male vs Female Purchases", fontweight = 'bold')
plt.show()
```

```python
gender = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount')
gender
```

```python
fig = sns.barplot(x = 'Gender', y = 'Amount', data = gender)

for bars in fig.containers:
    fig.bar_label(bars)

plt.show()
```

From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

#AGE

```python
a = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in a.containers:
    a.bar_label(bars)
```

```python
fig = sns.countplot(x = 'Age Group' , data =df ,hue = 'Gender')
for bars in fig.containers:
  fig.bar_label(bars)

plt.title("Buyers at different age group", fontweight = 'bold')
plt.show()
```

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by='Amount')
sales_age
```

```python
sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age)

plt.title('Total Amount vs Age Group', fontweight = 'bold')

plt.show()
```

# State

```python
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by='Orders')
sales_state
```

```python
sns.set(rc={'figure.figsize':(18,8)})
sns.barplot(x = 'State', y = 'Orders', data = sales_state)

plt.title('Orders from Top 10 states', fontweight = 'bold')

plt.show()
```

# Based on Occupation

```python
# Checking the order pattern based on Occupation
ax = sns.countplot(x = "Occupation", data = df)

sns.set(rc={"figure.figsize":(18,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Checking the pattern of Amount spent based on Occupation
sales_state = df.groupby(["Occupation"], as_index = False) ["Amount"].sum().sort_values(by = "Amount", ascending = False)

sns.set(rc={"figure.figsize":(18,5)})
ax = sns.barplot(x = "Occupation", y = "Amount", data = sales_state)
```

Based on the graphical data presented, it is evident that a significant proportion of buyers operate within the Information Technology, Healthcare, and Aviation industries.
